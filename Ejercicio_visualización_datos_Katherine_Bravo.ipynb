```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = "/content/drive/MyDrive/Colab Notebooks/PTCienciaDeDatos/PandasParaManipulaciónDeDatos/51000-0005_corregido.xlsx"
df = pd.read_excel(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df[df.columns[2:7]] = df[df.columns[2:7]].divide(1000000)
```

```python
df.columns = ("COD", "PRODUCT", "EXP_TON", "EXP_EUR", "EXP_USD", "IMP_TON", "IMP_EUR" , "IMP_USD")
```

```python
df.head()
```

```python
df1 = df.groupby(["PRODUCT"])[["EXP_TON","EXP_USD"]].sum().sort_values(by='EXP_TON', ascending=False).head(10)
df1
```

```python
import matplotlib.font_manager
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
import matplotlib.font_manager as fm
fm.FontManager().addfont('/usr/share/fonts/truetype/liberation/LiberationSans-BoldItalic.ttf')
```

```python
df2 = df.groupby(["PRODUCT"])[["EXP_TON","EXP_USD"]].sum().sort_values(by='EXP_USD', ascending=False).head(10)
df2
```

```python
fig_1 = plt.figure(dpi=200)
fig_1.subplots_adjust(top=1.25, wspace=0,hspace=-1)
ax1 = fig_1.add_subplot(1, 2, 1)
ax2 = fig_1.add_subplot(1, 2, 2)

fig_1.suptitle('Top 10 productos exportados de Alemania al mundo', fontsize=10, fontweight='bold', color='#1B4769')

colors1 = plt.cm.Blues(np.linspace(0.95, 0.38, len(df1)))

def autopct_with_values1(pct):
    total = sum(df1["EXP_TON"].values)
    val = int(round(pct * total / 100.0))
    return f"{pct:.1f}%\n ({val} MT)"

ax1.pie(df1["EXP_TON"].values, autopct=autopct_with_values1, colors=colors1, textprops={'fontsize': 3.5, 'color':'white'}, pctdistance=0.85)
ax1.add_artist(plt.Circle((0,0),0.7,fc='white'))
ax1.set_title('(2023, en mill. de toneladas)', fontsize=8.5, fontweight='bold', fontname='Liberation Sans', y=0.925, color='#1B4769')
ax1.legend(df1.index, loc='lower left', fontsize='4.5', bbox_to_anchor=(0, -0.31))
ax1.set_aspect('equal')

colors2 = plt.cm.Blues(np.linspace(0.95, 0.38, len(df1)))

def autopct_with_values2(pct):
    total = sum(df1["EXP_USD"].values)
    val = int(round(pct * total / 100.0))
    return f"{pct:.1f}%\n ({val} MT)"

ax2.pie(df2["EXP_USD"].values, autopct=autopct_with_values2, colors=colors2, textprops={'fontsize': 3.5, 'color':'white'}, pctdistance=0.85)
ax2.add_artist(plt.Circle((0,0),0.7,fc='white'))
ax2.set_title('(2023, en mill. de dolares)', fontsize=8.5, fontweight='bold', fontname='Liberation Sans', y=0.925, color='#1B4769')
ax2.legend(df1.index, loc='lower left', fontsize='4.5', bbox_to_anchor=(0, -0.31))
ax2.set_aspect('equal')

fig_1.text(0.25, 0.22, 'Source: © Statistisches Bundesamt (Destatis), 2024 | Stand: 11.03.2024 / 16:08:09\n (https://www.destatis.de/, Federal Statistical Office of Germany)', fontsize=5, color='#75A4C5')

plt.show()
```
